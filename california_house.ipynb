```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the Boston Housing dataset
california_housing = fetch_california_housing()
california_housing_df = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
california_housing_df['MEDV'] = california_housing.target  # Add the target variable (Median value of owner-occupied homes in $1000s)

# Display the first few rows of the dataset
california_housing_df.describe()
```

```python
california_housing_df.info()
```

```python
X = california_housing_df.iloc[:,:-1].values
y = california_housing_df.iloc[:,-1].values

print(X)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

regression = LinearRegression()

regression.fit(X_train, y_train)

pred = regression.predict(X_test)

print(len(X), len(y))
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,pred)
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
import numpy as np

svr_y = y.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, svr_y, test_size = 0.2, random_state = 0)


sc_X = StandardScaler()
sc_y = StandardScaler()
X_train_svr = sc_X.fit_transform(X_train)
y_train_svr = sc_y.fit_transform(y_train)

svr_regressor = SVR(kernel='rbf')

 # train_y.values.ravel()
svr_regressor.fit(X_train, np.ravel(y_train))

svr_pred = sc_y.inverse_transform(svr_regressor.predict(sc_X.transform(X_test)).reshape(-1,1))
np.set_printoptions(precision=2)
print(np.concatenate((svr_pred.reshape(len(svr_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
r2_score(y_test, svr_pred)
```

```python
rf_predict = rf_regressor.predict(X_test)
```
